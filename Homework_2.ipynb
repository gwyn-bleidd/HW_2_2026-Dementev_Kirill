```python
eng_stopwords = set(stopwords.words('english'))
morph = MorphAnalyzer()
patterns = "[0-9!#$%&'()*+,./:;<=>?@[\]^_`{|}~—\"\-]+"

def preprocess_text(text):
    text = text.lower()
    text = re.sub(patterns, ' ', text)
    tokens = []
    for token in text.split():
        if token and token not in eng_stopwords:
            token = token.strip()
            token = morph.normal_forms(token)[0]
            
            tokens.append(token)
    if len(tokens) > 2:
        return tokens
    return None


preprocessed_songs = []
song_names = []

for name, text in songs.items():
    clean_text = preprocess_text(text)
    preprocessed_songs.append(" ".join(clean_text))
    song_names.append(name)


df_songs = pd.DataFrame({
    'Song': song_names,
    'Processed song': preprocessed_songs
})


tfidf_vectorizer = TfidfVectorizer(
    analyzer='word',
    ngram_range=(1, 1),
    min_df=1,
    max_df=1.0,
    use_idf=True,
    smooth_idf=True,
    sublinear_tf=False
)

tfidf_matrix = tfidf_vectorizer.fit_transform(preprocessed_songs)
feature_names = tfidf_vectorizer.get_feature_names_out()

df_tfidf = pd.DataFrame(
    tfidf_matrix.toarray(),
    index=song_names,
    columns=feature_names
)

print("\nTF-IDF Матрица:")
print(df_tfidf.iloc[:, :20].round(4).to_string())
```

```python
# Создаем подложку для графиков
fig = plt.figure(figsize=(16, 10))

# гистограмма
ax2 = plt.subplot(3, 1, 1)
tfidf_values = tfidf_matrix.toarray().flatten()
tfidf_values = tfidf_values[tfidf_values > 0]
plt.hist(tfidf_values, bins=20, edgecolor='black', alpha=0.7, color='skyblue')
plt.title('Гистограмма распределения TF-IDF значений')
plt.xlabel('TF-IDF значение')
plt.ylabel('Частота')
plt.grid(True, alpha=0.3)

# ящик с усами
ax3 = plt.subplot(3, 1, 2)
df_tfidf_box = df_tfidf.T
bp = ax3.boxplot(
    [df_tfidf.iloc[i, :][df_tfidf.iloc[i, :] > 0].values for i in range(len(song_names))],
    labels=song_names,
    patch_artist=True
)
plt.title('Диаграмма "ящик с усами" распределения TF-IDF')
plt.ylabel('TF-IDF значение')
plt.grid(True, alpha=0.3)

colors = ['lightblue', 'lightgreen', 'lightcoral']
for patch, color in zip(bp['boxes'], colors):
    patch.set_facecolor(color)

# 10 слов с максимальным TF-IDF
ax4 = plt.subplot(3, 1, 3)
top_words = df_tfidf.mean().sort_values(ascending=False).head(10)
plt.barh(range(len(top_words)), top_words.values, color='lightgreen')
plt.yticks(range(len(top_words)), top_words.index)
plt.xlabel('Средний TF-IDF')
plt.title('Топ-10 слов с максимальным средним TF-IDF')
plt.gca().invert_yaxis()
plt.grid(True, alpha=0.3, axis='x')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
from gensim.models import Word2Vec
import time
from sklearn.metrics.pairwise import cosine_similarity


print("\n1) TF-IDF")
start_time = time.time()
tfidf_vectorizer = TfidfVectorizer()
tfidf_matrix = tfidf_vectorizer.fit_transform(preprocessed_songs)
tfidf_time = time.time() - start_time

print(f"time: {tfidf_time:.4f} сек")
print(f"matrix shape: {tfidf_matrix.shape}")
print(f"zero value percent: {1 - (tfidf_matrix.nnz / (tfidf_matrix.shape[0] * tfidf_matrix.shape[1])):.2%}")

print("\n2) Count Vectorizer")
start_time = time.time()
count_vectorizer = CountVectorizer()
count_matrix = count_vectorizer.fit_transform(preprocessed_songs)
count_time = time.time() - start_time

print(f"time: {count_time:.4f} сек")
print(f"matrix shape: {count_matrix.shape}")
print(f"zero value percent: {1 - (count_matrix.nnz / (count_matrix.shape[0] * count_matrix.shape[1])):.2%}")

print("\n3) Word2Vec")

tokenized_songs = [song.split() for song in preprocessed_songs]

start_time = time.time()
w2v_model = Word2Vec(sentences=tokenized_songs, 
                     vector_size=100,
                     window=5,
                     min_count=1,   
                     workers=1)
w2v_train_time = time.time() - start_time

def get_doc_vector(tokens, model, vector_size):
    word_vectors = []
    for token in tokens:
        if token in model.wv:
            word_vectors.append(model.wv[token])

    if word_vectors:
        return np.mean(word_vectors, axis=0)
    else:
        return np.zeros(vector_size)

start_time = time.time()
doc_vectors = []
for tokens in tokenized_songs:
    doc_vector = get_doc_vector(tokens, w2v_model, 100)
    doc_vectors.append(doc_vector)
w2v_inference_time = time.time() - start_time

doc_vectors = np.array(doc_vectors)
print(f"time: {w2v_train_time:.4f} сек")
print(f"time of vectors: {w2v_inference_time:.4f} сек")
print(f"matrix shape: {doc_vectors.shape}")
print(f"zero value percent: {1 - (np.count_nonzero(doc_vectors) / doc_vectors.size):.2%}")

print("\nВычислительная сложность:")

analysis_data = {
    'Метод': ['TF-IDF', 'CountVectorizer', 'Word2Vec'],
    'Время (сек)': [tfidf_time, count_time, w2v_train_time + w2v_inference_time],
    'Размерность': [f"{tfidf_matrix.shape}", f"{count_matrix.shape}", f"{doc_vectors.shape}"]
}

df_complexity = pd.DataFrame(analysis_data)
print(df_complexity.to_string(index=False))


print("\n\nКачество представления:")

def plot_similarity_matrix(matrix, method_name):
    similarity = cosine_similarity(matrix)
    
    plt.figure(figsize=(8, 6))
    sns.heatmap(similarity, annot=True, fmt='.3f', 
                xticklabels=song_names,
                yticklabels=song_names,
                vmin=0, vmax=1)
    plt.title(f'Матрица сходства песен - {method_name}')
    plt.tight_layout()
    plt.show()
    
    return similarity


print("\nМатрицы косинусного сходства между песнями:")

print("\nTF-IDF:")
tfidf_sim = plot_similarity_matrix(tfidf_matrix, 'TF-IDF')

print("\nCountVectorizer:")
count_sim = plot_similarity_matrix(count_matrix, 'Count Vectorizer')

print("\nWord2Vec:")
w2v_sim = plot_similarity_matrix(doc_vectors, 'Word2Vec')

print("\nАнализ качества представления:")
quality_data = []

# TF-IDF
tfidf_unique = np.sum(tfidf_matrix.toarray() > 0, axis=1)
quality_data.append({
    'Метод': 'TF-IDF',
    'Ср. кол-во значимых признаков': f"{np.mean(tfidf_unique):.1f}",
    'Диапазон значений': f"[{tfidf_matrix.min():.3f}, {tfidf_matrix.max():.3f}]"
})

# CountVectorizer
count_unique = np.sum(count_matrix.toarray() > 0, axis=1)
quality_data.append({
    'Метод': 'CountVectorizer',
    'Ср. кол-во значимых признаков': f"{np.mean(count_unique):.1f}",
    'Диапазон значений': f"[{count_matrix.min()}, {count_matrix.max()}]"
})

# Word2Vec
w2v_range = f"[{doc_vectors.min():.3f}, {doc_vectors.max():.3f}]"
quality_data.append({
    'Метод': 'Word2Vec',
    'Ср. кол-во значимых признаков': f"{doc_vectors.shape[1]} (все)",
    'Диапазон значений': w2v_range
})

df_quality = pd.DataFrame(quality_data)
print(df_quality.to_string(index=False))

print("\nИнтерпретируемость:")

interpretability_data = {
    'Метод': ['TF-IDF', 'CountVectorizer', 'Word2Vec'],
    'Интерпретируемость': ['Высокая', 'Средняя', 'Низкая'],
    'Объяснение': [
        'Можно сразу сказать, какие слова важны и насколько',
        'Показывает частоты слов',
        'Векторы слов неинтерпретируемы'
    ]
}

df_interpret = pd.DataFrame(interpretability_data)
print(df_interpret.to_string(index=False))


print("Итого:")
print("""Вычислительная сложность:
   - CountVectorizer самый быстрый, так как просто считает частоты
   - TF-IDF немного медленнее из-за вычисления IDF
   - Word2Vec требует много времени на обучение
""")

print("""Качество представления:
   - TF-IDF лучше отражает важность слов в контексте коллекции
   - CountVectorizer простой, но не взвешивает важность терминов
   - Word2Vec понимает семантические связи
      """)
print("""Интерпретируемость:
   - TF-IDF: можно объяснить, почему слово важно
   - CountVectorizer: понятно, но не информативно
   - Word2Vec: векторы неинтерпретируемы
""")
```

```python
# Статистический анализ


stats_data = []
for i, song_name in enumerate(song_names):
    song_freq = count_matrix[i].toarray().flatten()
    song_freq_nonzero = song_freq[song_freq > 0]
    
    stats_data.append({
        'Песня': song_name,
        'Уникальных слов': len(song_freq_nonzero),
        'Всего слов': int(song_freq.sum()),
        'Средняя частота': np.mean(song_freq_nonzero),
        'Макс. частота': np.max(song_freq_nonzero),
        'Мин. частота': np.min(song_freq_nonzero),
        'Медианная частота': np.median(song_freq_nonzero)
    })

df_stats = pd.DataFrame(stats_data)
print("\nСтатистика по песням:")
print(df_stats.round(2).to_string(index=False))
```
